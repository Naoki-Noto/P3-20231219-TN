```python
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from mordred import Calculator, descriptors
```

```python
df_cate = pd.read_csv('Catechol_smiles.csv', index_col=0)
df_cate
```

```python
calc = Calculator(descriptors, ignore_3D = True)
df_1 = calc.pandas([Chem.MolFromSmiles(smile) for smile in df_cate['SMILES']])
display(df_1)
```

```python
df_Cate_desc = df_1.loc[df_1.index.repeat(4)].reset_index(drop=True)
print(df_Cate_desc.info())
df_Cate_desc.head(10)
```

```python
df_P = pd.read_csv('P_smiles.csv', index_col=0)
df_P
```

```python
calc = Calculator(descriptors, ignore_3D = True)
df_2 = calc.pandas([Chem.MolFromSmiles(smile) for smile in df_P['SMILES']])
display(df_2)
```

```python
df_P_desc=pd.concat([df_2]*17,ignore_index=True)
print(df_P_desc.info())
df_P_desc.head(10)
```

```python
mordred_desc=pd.concat([df_Cate_desc, df_P_desc], axis=1)
mordred_desc.to_csv('mordred_desc.csv')
mordred_desc.head(10)
```
