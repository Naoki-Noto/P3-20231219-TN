```python
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit_installer.fingerprints import Fingerprinter
```

```python
df_cate = pd.read_csv('Catechol_smiles.csv', index_col=0)
df_cate
```

```python
fingerprinter = Fingerprinter()
fp_df_cate = pd.DataFrame([vec for vec in fingerprinter.transform(df_cate['SMILES'], fp_type='Morgan2(1024bits)')])

print(fp_df_cate.shape)
fp_df_cate.head()
```

```python
df_Cate_desc = fp_df_cate.loc[fp_df_cate.index.repeat(4)].reset_index(drop=True)
print(df_Cate_desc.info())
df_Cate_desc.head(10)
```

```python
df_P = pd.read_csv('P_smiles.csv', index_col=0)
df_P
```

```python
fingerprinter = Fingerprinter()
fp_df_P = pd.DataFrame([vec for vec in fingerprinter.transform(df_P['SMILES'], fp_type='Morgan2(1024bits)')])

print(fp_df_P.shape)
fp_df_P.head()
```

```python
df_P_desc=pd.concat([fp_df_P]*17,ignore_index=True)
print(df_P_desc.info())
df_P_desc.head(10)
```

```python
Morgan_desc=pd.concat([df_Cate_desc, df_P_desc], axis=1)
Morgan_desc.to_csv('Morgan_desc.csv')
print(Morgan_desc.info())
Morgan_desc.head(10)
```
