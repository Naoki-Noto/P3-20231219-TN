```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import PandasTools
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Descriptors
from sklearn.decomposition import PCA 
```

```python
df_cate = pd.read_csv('Catechol_smiles.csv', index_col=0)
df_cate
```

```python
cate_list=[Chem.MolFromSmiles(smiles) for smiles in df_cate['SMILES']]
cate_img = Draw.MolsToGridImage(cate_list,molsPerRow=5)
cate_img
```

```python
PandasTools.AddMoleculeColumnToFrame(df_cate,'SMILES')
for i,j in Descriptors.descList:
    df_cate[i] = df_cate.ROMol.map(j)
df_cate = df_cate.add_prefix('Cate_')
df_cate.to_csv('./component/Cate_rdkit.csv')
df_cate
```

```python
df_Cate_dr=df_cate.drop(columns=['Cate_SMILES','Cate_ROMol'])
df_Cate_desc = df_Cate_dr.loc[df_Cate_dr.index.repeat(4)].reset_index(drop=True)
print(df_Cate_desc.info())
df_Cate_desc.head(10)
```

```python
#Preparing
# delete variables with zero variance
Var0Variable = np.where( df_cate.var(axis=0) == 0 )
if len(Var0Variable[0]) == 0:
    print( "There is no variable with zero variance." )
    print( "" )
else:
    print( "There were {0} variables with zero variance".format(len(Var0Variable[0])))
    print( "the number of variable is: {0}".format(Var0Variable[0]) )
    print( "remove them" )
    print( "" )
    df_var0_cate = df_cate.drop(df_cate.columns[Var0Variable], axis=1)

#removal of columns which data has all 0
def remove_all_zero_col(df):
    df = df.copy()
    for col in df.columns:
        if (df[col] == 0).all():
            df.drop(col, axis=1, inplace=True)
    return df
df_var0_cate = remove_all_zero_col(df_var0_cate)
df_var0_cate.info()
```

```python
X_cate = df_var0_cate.iloc[:, 2:]
X_cate.info()
```

```python
#autoscale
a_X_cate = (X_cate - X_cate.mean(axis=0)) / X_cate.std(axis=0,ddof=0)
a_X_cate = a_X_cate.dropna(how='all', axis=1)
```

```python
#n_components = number of df columns
pca = PCA()
pca.fit(a_X_cate)

#PCA
pca_cor_cate = pca.transform(a_X_cate)
pca_X_cate =pd.DataFrame(pca_cor_cate)

pca_cols_cate = ["rdkit_catePC{}".format(x + 1) for x in range(len(pca_X_cate.columns))]
a_cateX_pca = pd.DataFrame(pca_cor_cate, columns = pca_cols_cate)
a_cateX_pca
```

```python
a_cateX_pca_desc = a_cateX_pca.loc[a_cateX_pca.index.repeat(4)].reset_index(drop=True)
print(a_cateX_pca_desc.info())
a_cateX_pca_desc.head(10)
```

```python
#contribution
df_exp_var_ratio_cate = pd.DataFrame([pca.explained_variance_ratio_],columns=pca_cols_cate)
print(df_exp_var_ratio_cate.head())

# sum of explained variance ratio
cum_exp_var_ratio_cate = np.hstack([0, pca.explained_variance_ratio_]).cumsum()
plt.plot(cum_exp_var_ratio_cate, 'D-')
plt.title('Cumulative contribution rate (cate_rdkit)')
plt.xticks(range(18))
plt.yticks(np.arange(0,1.05,0.1))
plt.xlabel(r"the numbeer of pca descriptors",color='k', size=10)
plt.ylabel(r"$Cumulative contribution rate$",color='k', size=10)
plt.grid()
plt.show()

a_cateX_pca.to_csv('./component/pcaRDKit_Cate.csv')
a_cateX_pca_desc.to_csv('./component/pcaRDKit_Cate_descriptor.csv')
df_exp_var_ratio_cate.to_csv('./component/contribution_ratio_Cate.csv')
```
